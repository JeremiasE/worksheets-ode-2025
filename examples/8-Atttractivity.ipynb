```python
import IPython.display as disp
import sympy
import vector_fields
from sympy import init_printing
init_printing() 
```

```python
(t,x,y)=sympy.var("t,x,y")
r=sympy.sqrt(x**2+y**2)
```

# Eine Differentialgleichung mit instabiler aber attraktiver Ruhelage

```python
f1 = (-y/2*(1-x/r)+x*(1-r),(x/2*(1-x/r)+y*(1-r)))
```

```python
vector_fields.plot_streamlines(f1,(x,-6,6),(y,-4,4)); 
```

```python
vector_fields.field_player(f1)
```

# Eine Differentialgleichung mit instabiler und global attraktiver Ruhelage

```python
f2 = ((x**2*(y-x)+y**5)/(r**2*(1+r**4)),(y**2*(y-2*x))/(r**2*(1+r**4)))
vector_fields.plot_streamlines(f2,(x,-6,6),(y,-4,4));
```

```python
vector_fields.field_player(f2)
```
